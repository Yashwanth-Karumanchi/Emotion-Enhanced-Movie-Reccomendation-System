```python
import datasets as ds
import tensorflow as tf
import numpy as np
dataset=ds.load_dataset('dair-ai/emotion')
```

```python
print(type(dataset))
train=dataset['train']
test=dataset['test']
validate=dataset['validation']
print(len(test))
```

```python
def get_tweets(data):
  tweets=[x['text'] for x in data]
  labels=[x['label'] for x in data]
  return tweets,labels
```

```python
tweets,labels = get_tweets(train)
print(type(tweets))
```

```python
from tensorflow.keras.preprocessing.text import  Tokenizer
tokenizer=Tokenizer(num_words=1000,oov_token='<UNK>')
tokenizer.fit_on_texts(tweets)
tokenizer.texts_to_sequences([tweets[0]])
```

```python
maxlen=50
from tensorflow.keras.preprocessing.sequence import pad_sequences
def get_sequences(tokenizer,tweets):
  sequences=tokenizer.texts_to_sequences(tweets)
  padded=pad_sequences(sequences,truncating='post',padding='post',maxlen=maxlen)
  return padded
```

```python
padded_train_seq = get_sequences(tokenizer,tweets)
padded_train_seq[0]
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Embedding(1000,16,input_length=maxlen),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(20,return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(20)),
    tf.keras.layers.Dense(6,activation='softmax')
])
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
val_tweets,val_labels=get_tweets(validate)
val_seq=get_sequences(tokenizer,val_tweets)
print(val_tweets[1],val_labels[1])
```

```python
h=model.fit(
    padded_train_seq,np.array(labels),
    validation_data=(np.array(val_seq),np.array(val_labels)),
    epochs=20,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',patience=2)
    ]
)
```

```python
string=input("enter a string ")
string=get_sequences(tokenizer,[string])
print(type(string))
p=model.predict(string)[0]
print(p)
res=np.argmax(p).astype('uint8')
print(res)
print("mood detected is ....")
if res==0:
  print("sadness")
elif res==1:
  print("joy")
elif res==2:
  print("love")
elif res==3:
  print("anger")
elif res==4:
  print("fear")
else:
  print("surprise")
```

```python
temp=["Drama","Horror","Romance","Biography","Romance","Animation"]
search_for_genre=temp[res]
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_movies=pd.read_csv("imdb_top_1000.csv")
```

```python
# print(df_movies.head(10))
```

```python

tfidf = TfidfVectorizer(stop_words='english')


tfidf_matrix = tfidf.fit_transform(df_movies['Genre'])


cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)



def recommend_movies(genre):
    indices = df_movies[df_movies['Genre'] == genre].index
    genre_cosine_sim = cosine_sim[indices].sum(axis=0)
    top_indices = genre_cosine_sim.argsort()[::-1][:10]
    return df_movies.loc[top_indices]['Series_Title']
print("suggested genre is ",search_for_genre)
recommended_movies = recommend_movies(search_for_genre)
print(recommended_movies)
```
