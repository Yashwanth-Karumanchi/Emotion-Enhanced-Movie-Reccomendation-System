```python
model_instance=tf.keras.models.load_model("model_instance_20epochs")
```

```python
string=input("enter a string ")
string=get_sequences(tokenizer,[string])
print(type(string))
p=model_instance.predict(string)[0]
print(p)
res=np.argmax(p).astype('uint8')
print(res)
if res==0:
  print("sadness")
elif res==1:
  print("joy")
elif res==2:
  print("love")
elif res==3:
  print("anger")
elif res==4:
  print("fear")
else:
  print("surprise")
```

```python
temp=["Drama","Horror","Romance","Biography","Romance","Animation"]
search_for_genre=temp[res]
```

```python
df_movies=pd.read_csv("imdb_top_1000.csv")
```

```python

tfidf = TfidfVectorizer(stop_words='english')


tfidf_matrix = tfidf.fit_transform(df_movies['Genre'])


cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)



def recommend_movies(genre):
    indices = df_movies[df_movies['Genre'] == genre].index
    genre_cosine_sim = cosine_sim[indices].sum(axis=0)
    top_indices = genre_cosine_sim.argsort()[::-1][:10]
    return df_movies.loc[top_indices]['Genre']
print("suggested genre is ",search_for_genre)
recommended_movies = recommend_movies(search_for_genre)
print(type(recommended_movies))
```
